# Ejercicios con la clase Nodo

- Crear nodo
  - imprimir el valor del nodo
- Crear dos nodos 10, 20
  - imprimir el valor del primero
  - imprimir el valor del segundo
  - transversal
- Lo mismo pero con 4 nodos 10, 20, 30, 40
  - imprimir el valor del primero
  - imprimir el valor del ultimo
  - transversal
  - insertar el nodo 25, al final

```python
class NodoDoble:
    def __init__( self , value , siguiente = None , previo = None ):
        self.data = value
        self.next = siguiente
        self.prev = previo
```

```python
# Ejercicio 1
print("Ejercicio 1")
head = NodoDoble(10)
tail = head
print( head.data )
print( tail.data )

# Ejercicio 2
print()
print("Ejercicio 2")
head = NodoDoble( 10 , NodoDoble(20))
head.next.prev = head
tail = head.next
print("El primero:" , head.data)
print("El primero:" , tail.prev.data)

print("El primero:" , head.next.data)
print("El primero:" , tail.data)

# Ejercicio 3
print()
print("Ejercio 3")
head = NodoDoble(10 , NodoDoble(20 , NodoDoble(30 , NodoDoble(40))))
tail = head.next.next.next
head.next.prev = head
head.next.next.prev = head.next 
tail.prev = head.next.next

print("El primero:" , head.data)
print("El primero:" , tail.prev.prev.prev.data)

print("El último:" , tail.data)
print("El último:" , head.next.next.next.data)
print()

#Transversal
curr_node = head
while curr_node != None:
    print(f"|{curr_node.data}|-->" , end="")
    curr_node = curr_node.next
print("")
print()

#Reversal transversal
curr_node = tail
while curr_node != None:
    print(f"|{curr_node.data}|-->" , end="")
    curr_node = curr_node.prev
print("")
print()
```

```python
class DoubleLinkedList:
    def __init__( self  ):
        self.head = None
        self.tail = None
        self.size = 0

    def get_size( self ):
        return self.size

    def is_empty( self ): # true si la lista está vacía
        return self.size == 0

    def append( self , value ):
        if self.is_empty():
            self.head = NodoDoble( value)
            self.tail = self.head
        else:
            self.tail.next = NodoDoble(value , None , self.tail )
            self.tail = self.tail.next
        self.size += 1

    def transversal( self ):
        curr_node = self.head
        while curr_node != None :
            print(f"| {curr_node.data} | --> " , end="")
            curr_node = curr_node.next
        print("")

    def remove_from_head( self , value  ):
        curr_node = self.head
        while curr_node.data != value:
            curr_node = curr_node.next

        if curr_node.data == value:
            print("Encontrado",curr_node.data)
            curr_node.prev.next = curr_node.next
            curr_node.next.prev = curr_node.prev
            print("", curr_node.prev.data)
        self.size -= 1

    def find_from_tail( self , value ):
        curr_node = self.tail
        contador = 0
        while curr_node.data != value:
            contador += 1
            curr_node = curr_node.prev
        
        if curr_node.data == value:
            print(f"El nodo está en la posición:{ contador + 1 } desde tail")


    def find_from_head( self , value ):
        curr_node = self.head
        contador = 0
        while curr_node.data != value:
            contador += 1
            curr_node = curr_node.next

        if curr_node.data == value:
            print(f"El Nodo está en la posición:{ contador + 1 } desde head")


    def remove_from_tail( self , value ):
        curr_node = self.tail
        while curr_node.data != value:
            curr_node = curr_node.prev
        
        if curr_node.data == value:
            curr_node.next.prev = curr_node.prev
            curr_node.prev.next = curr_node.next
        self.size += 1

    def reverse_transversal( self ):
        curr_node = self.head
        while curr_node != None :
            print(f"| {curr_node.data} | --> " , end="")
            curr_node = curr_node.next
        print("")
```

```python
ldl = DoubleLinkedList()
ldl.append(10)
ldl.append(20)
ldl.transversal()
print(f"Esta vacía? { ldl.is_empty() }")

ldl = DoubleLinkedList()
ldl.append(10)
ldl.append(20)
ldl.append(30)
ldl.append(40)
ldl.append(50)
ldl.append(60)
ldl.transversal()
print()
print(f"¿La lista está vacía?: { ldl.is_empty() }")
print()
print("remove_from_head")
ldl.remove_from_head(20)
ldl.transversal()
print()
print("remove_from_tail")
ldl.remove_from_tail(30)
ldl.transversal()
print()
print(f"La Lista Doblemente Ligada tiene { ldl.get_size() } elementos")
print()
ldl.find_from_tail(10)
print()
ldl.find_from_head(50)
print()
print("reverse_transversal:")
ldl.reverse_transversal()
print()
```

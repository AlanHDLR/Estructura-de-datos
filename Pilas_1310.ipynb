```python
class StackADT:
    def __init__( self ):
        self.__data = list()

    # regresa true si la lista esta vacia 
    def is_empty( self ):
        return len( self.__data ) == 0

    def get_lenght( self ):
        return len( self.__data )

    def pop( self ):
        if self.is_empty():
            print("Pila vacia, nada que sacar")
        else:
            return self.__data.pop()

    def peek( self ):
        if self.is_empty():
            print("Pila vacia, nada que sacar")
        else:
            return self.__data[ -1 ]

    def push( self , value ):
        self.__data.append( value )

    def to_string( self ):
        for item in self.__data[::-1]:
            print(f" | { item } | ")
```

```python
pila = StackADT()
print( "¿Estas vacia? =", pila.is_empty() )
print( "# elementos:" , pila.get_lenght() )
print()
pila.push(10)
print( "¿Estas vacia? =", pila.is_empty() )
print( "# elementos:" , pila.get_lenght() )
print()
pila.push(20)
pila.push(30)
pila.push(40)
pila.to_string()
print( pila.pop() ) # 40
print()
print( pila.peek() ) # 30
pila.to_string()
```
